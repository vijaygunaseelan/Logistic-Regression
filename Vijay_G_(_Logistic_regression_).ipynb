# 4.Build  the Best model?

```python
x_ind11=x_ind.drop("Class",axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
x_train,x_test,y_train,y_test=train_test_split(x_ind11,y_dep,test_size=0.2,random_state=3)
```

```python
import statsmodels.api as stm
modelstat=stm.Logit(y_train,x_train).fit()
modelstat.summary2()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_train=LogisticRegression()
```

```python
model_train.fit(x_train,y_train)
```

```python
y_pred=model_train.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,plot_confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
acc=model_train.score(x_test,y_test)
```

```python
acc
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc=roc_auc_score(y_test,y_pred)
```

```python
fpr,tpr,thresholds=roc_curve(y_test,model_train.predict_proba(x_test)[:,1])
```

```python
plt.plot(fpr,tpr,logit_roc_auc)
```

```python
z=LogisticRegression(class_weight="balanced")
z.fit(x_train,y_train)
THRESHOLD=0.48
```

```python
pred=np.where(z.predict_proba(x_test)[:,1]>THRESHOLD,1,0)
```

```python
pred
```

```python
pd.DataFrame(data=[accuracy_score(y_test,pred)],index=["ROC_prediction"])
```
